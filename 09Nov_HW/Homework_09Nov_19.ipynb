```
import string
from string import digits

def preprocess(x):
    ext_punct = string.punctuation + '«»—'
    x = x.translate(str.maketrans('','',ext_punct)) # удаление пунктуации
    x = x.translate(str.maketrans('','',string.digits)) # удаление цифр
    x1 = x.lower().split()
    return(x1)

with open ('city_smells.txt','r', encoding = 'utf-8') as smells:
    smells = smells.read()
       
    # определение средней длины слова в тексте:
    
    lst_smells = preprocess(smells)
    a = 0
    for i in lst_smells:
        a+=len(i)
    med_words_len = a/len(lst_smells)
    with open ('med_words_len.txt', 'w') as res_med_words_len: # записываем результат в новый файл 
        res_med_words_len.write("Средняя длина слова в тексте - "+ str(round(med_words_len,3)))
    
   

    

```

```
# определить среднюю длину предложения в тексте
with open ('city_smells.txt','r', encoding = 'utf-8') as smells:
    smells = smells.read()    
    
    import re
        
    a = re.sub('[\.\?\!] ','\.\n', smells)
    a = a.split('\n') # разбиение на предложения
    cnt = 0 # счетчик символов 
    
    for i in a:
        cnt+=len(i)
    med_len_sent = cnt/len(a)
    with open ('med_len_sent.txt', 'w') as res_med_len_sent: # записываем результат в новый файл 
        res_med_len_sent.write("Средняя длина предложения в тексте: "+ str(round(med_len_sent,3)))
    
```
