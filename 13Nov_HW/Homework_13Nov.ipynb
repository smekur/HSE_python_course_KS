**Задание 1** - нахождение адреса


```
# Адрес в данном файле представляет собой три строки, первая из которых начинается с цифр (больше подобной структуры в этом файле нет).
# Можно найти строку, начинающуюся с цифр, а далее взять две следующие за ней строки и сформировать таким образом адрес.

import re

with open ('addresses.txt','r', encoding = 'utf-8') as addr:
    addr = addr.readlines() # читаем файл, создаем список по строкам
    a = [] # создаем пустой список для сохранения индексов нужных нам строчек - начала адреса
    for i in addr: 
        
        res = re.findall('^(\d+)',i) # находим подходящие строки
        if res == []:
            continue
        else:
              
            b = [j for j,x in enumerate(addr) if x == i] # ищем индексы всех вхождений, соответствующих запросу
            a+=b # запоминаем индексы
    
    a = set(a) # избавляемся от индексов-дубликатов 
    
    res_addr = [] # создаем список для хранения получившихся адресов
    for j in a:
        c = [] # промежуточный список для объединения строчек адреса
        c.append(addr[j].strip())
        c.append(addr[j+1].strip()) 
        c.append(addr[j+2].strip())
        
        c = ", ".join(c)
        '''
        # при необходимости, можно отбросить одинаковые адреса:
        if c not in res_addr:
            res_addr.append(c)
        else:
            continue
        '''
        res_addr.append(c)
            
    res_output = "; ".join(res_addr)
    with open ('res_address.txt', 'w') as res_address: # записываем результат в новый файл 
        res_address.write(res_output)
```

**Задание 3** - замените все случаи употребления строки "язык" на "шашлык"



```
import re

with open ('ея.txt','r', encoding = 'utf-8') as nat_lang:
    nat_lang = nat_lang.read()
    res_shashlyk = re.sub('яз(ы|ы́)к', 'шашлык', nat_lang)
    with open ('shashlyk.txt', 'w') as shashlyk: # записываем результат в новый файл 
        shashlyk.write(res_shashlyk)
```
